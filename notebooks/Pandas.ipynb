### I. What is Pandas?

Pandas is a fast, powerful, flexible and easy to use open source `data analysis and manipulation tool`.

And to analyze and manipulate our data using Pandas, we need to create a Dataframe.
`DataFrame` is a 2-dimensional data structure providing a flexible and intuitive way to store and work with data..

### X. How to handle missing data, duplicate data, categorical data, wrong data and wrong format?

<br>**Missing Data:**

```python
# Check how many null values each column has
df_3.isnull().sum()
```

```python
# Drop null values
df_3.dropna(inplace=True)
```

```python
# Compute the mean of a column
carat_mean = df_3.iloc[:, 0].mean()
# Fill the null values with computed mean value
df_3.iloc[:, 0].fillna(carat_mean, inplace=True)
```

<br>**Duplicate Data:**

```python
# Check the presence of duplicate rows
df_3.duplicated().values.any()
```

```python
# Drop duplicate rows
df_3.drop_duplicates(keep='first', inplace=True)
```

<br>**Categorical Data:**

```python
# Check the columns that have object dtypes
df_3.select_dtypes(include=['object']).columns
```

```python
# One-Hot Encoding
ohe_df = pd.get_dummies(df_3, dtype=int)

ohe_df
```

```python
# Create dictionary for each categorical features
cut_dict = {"Fair": 1, "Good": 2, "Very Good": 3, "Premium": 4, "Ideal": 5}
color_dict = {"J": 1,"I": 2,"H": 3,"G": 4,"F": 5,"E": 6,"D": 7}
clarity_dict = {"I1": 1, "SI2": 2, "SI1": 3, "VS2": 4, "VS1": 5, "VVS2": 6, "VVS1": 7, "IF": 8}

# Map the 3 dictionaries
df_3['cut'] = df_3['cut'].map(cut_dict)
df_3['color'] = df_3['color'].map(color_dict)
df_3['clarity'] = df_3['clarity'].map(clarity_dict)

df_3
```

<br>**Wrong Data:**

```python
df_3.loc[0, 'cut'] = 'ideall'
```

```python
df_3.iloc[:, 1].value_counts()
```

```python
# Replace wrong values
df_3.replace('ideall', '5', inplace=True)
```

```python
df_3.iloc[:, 1].value_counts()
```

<br>**Wrong Format:**

```python
# Convert to correct data type/format
df_3 = df_3.astype({'cut': 'int64'})
```

```python
# Check if its now in the correct format
df_3.iloc[:, 1].dtypes
```

```python
# Check if its now in the correct format
df_3.iloc[:, 1].value_counts()
```
