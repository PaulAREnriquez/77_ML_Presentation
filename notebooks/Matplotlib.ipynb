### I. What is Matplotlib?

Matplotlib is a comprehensive library for `creating` static, animated, and interactive `visualizations` in Python. 

`Figure` is where matplotlib graphs the data. Also, considered as the Top level artist, which holds all plot elements.

### II. How to get started?

1. Install using PIP command: `pip install matplotlib`

2. Import to use its functions: `import matplotlib.pyplot as plt`

```python
# Import Package
import matplotlib.pyplot as plt

import matplotlib
import pandas as pd
```

```python
# Check Version
matplotlib.__version__
```

### III. How to create and view a visualization?

```python
# Read data
df = pd.read_csv('../datasets/diamond_prices.csv')
```

```python
depth = df['depth']
price = df['price']
cut = df['cut'].value_counts()
```

```python
# Create a single Visualization

# Create a Figure
fig = plt.figure()

# Create axes by instantiating an instance of a figure class using subplots
ax = fig.subplots()

# Call a drawing method
ax.scatter(depth, price, color='green')
```

```python
# Create multiple Visualizations

# Create a Figure and axes
fig2, axs = plt.subplots(2, 2, figsize=(10, 8))

# Call drawing methods
axs[0,0].plot(depth, price)
axs[0,1].hist(price, bins=5, color='red')
axs[1,0].boxplot(depth)
axs[1,1].pie(cut, frame=True)
```

### IV. How to save a visualization?

```python
# Save a visualization
fig.savefig('./figures/sample_figure.png')
fig2.savefig('./figures/sample_figure2.png')
```

### VI. How to style a visualization?

```python
fig = plt.figure()
ax = fig.subplots()
ax.hist(price, bins=5, color='orange')

# Set figure title
fig.suptitle('Figure Title', color='tab:red')

# Set axes title
ax.set_title('Axes Title', color='g')

# Set xlabel title
ax.set_xlabel('x-axis label', color='#A020F0')

# Set ylabel title
ax.set_ylabel('y-axis label', color=(0, 0, 1))

# Set legend
ax.legend(['Sample Legend'], loc='upper right')

# Set grid
ax.grid()
```
